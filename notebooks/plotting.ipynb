```python
from dataclasses import dataclass
import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', family='D2Coding')
plt.rc('image', cmap='Pastel1')

@dataclass
class PieNode:
    value: int
    label: str
    color: str
```

```python
n_all_images = 141931
n_kdj_todo = 12712
n_kdj_did = n_kdj_todo - 5718
n_pdb_todo = 12579

plt.pie([n_kdj_todo-n_kdj_did, n_kdj_did, n_pdb_todo, n_all_images-n_kdj_todo-n_pdb_todo],
        radius=1,
        labels=[
            f'{n_kdj_todo-n_kdj_did}/{n_kdj_todo} (김동주-진행 중)',
            f'{n_kdj_did}/{n_kdj_todo} (김동주-작업 완료)',
            '',
            '',
        ],
        colors=[
            'salmon',
            'mistyrose',
            'white',
            'white'
        ],
        wedgeprops={'width': 0.2, 'edgecolor': 'w'})
plt.pie([n_kdj_todo, n_pdb_todo, n_all_images-n_kdj_todo-n_pdb_todo],
        radius=0.8,
        labels=[
            '',
            f'{n_pdb_todo} (박다빈)',
            f'{n_all_images-n_kdj_todo-n_pdb_todo} (라벨링 대상 아님)'
        ],
        colors=[
            'mistyrose',
            'powderblue',
            'lavender'
        ],
        wedgeprops={'width': 0.4, 'edgecolor': 'w'})
plt.title('Product-10K 데이터 라벨링 분배량 (이미지 장수, 2인 작업 기준)')
```

```python
n_images = [1143, 1144, 1144, 1143, 1144, n_kdj_did]
names = ['1143 (김동주)', '1144 (배상혁)', '1144 (배희재)', '1143 (윤은수)', '1144 (전경민)', f'{n_kdj_did} (라벨링 완료)']

plt.pie(n_images,
        radius=1,
        labels=names,
        colors=[
            'lightcoral',
            'indianred',
            'firebrick',
            'maroon',
            'darkred',
            'steelblue',
        ],
        wedgeprops={'width': 0.4, 'edgecolor': 'w'})
plt.title('1차 분배량 (김동주-진행 중 5,718장 재분배)')
```
